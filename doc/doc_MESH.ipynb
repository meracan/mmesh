## Examples

```python
# Using SLF object to create a simple grid
slf=SLF() # Create default grid
path="example1.slf"
slf.write(path) # Write default grid to file

_m=MESH.read(path)

# Plot
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(8, 8))
fig.tight_layout()
_m.plot(axe=axes)
_m.boundaries.plot(axe=axes,showPoints=True,pointStyle={"c":"black"},polygonStyle={'facecolor':(0,0,0,0)})
MESH.delete(path)
```

```python
# Using SLF object to create a simple grid
slf="example1.slf"
msh="example1.msh"

SLF().write(slf) # Write default grid to file
m1=MESH.read(slf).write(msh)
m2=MESH.read(msh)

# Plot
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(8, 8))
fig.tight_layout()
m2.plot(axe=axes)
m2.boundaries.plot(axe=axes,showPoints=True,pointStyle={"c":"black"},polygonStyle={'facecolor':(0,0,0,0)})
MESH.delete(slf)
MESH.delete(msh)
pass
```
